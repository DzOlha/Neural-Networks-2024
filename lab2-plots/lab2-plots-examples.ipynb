```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]);
plt.show();
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]);
plt.title('First plot');
plt.xlabel('x');
plt.ylabel('y');
plt.show();
```

```python
plt.figure(figsize=(15, 5));
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]);
plt.title('First plot');
plt.xlabel('x');
plt.ylabel('y');
plt.show();
```

```python
x = np.arange(1, 5);
y = x**3;

# rows, columns, index
plt.subplot(1, 2, 1);
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]);
plt.title('First Subplot');

plt.subplot(1, 2, 2);
plt.plot(x, y);
plt.title("Second Subplot");

plt.show();
```

```python
x = np.arange(1, 5);
y = x**3;

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "go", x, y, "r^");
plt.title("Merged Plot");
plt.xlabel('x');
plt.ylabel('y');
plt.show();
```

```python
div=["div_a", "div_b", "div_c", "div_d"];
div_average = [20, 60, 30, 50];
plt.bar(div, div_average, color = 'lightblue');
plt.title('Bar Graph');
plt.xlabel('x');
plt.ylabel('y');
plt.show();
```

```python
firms = ["firm1", "firm2", "firm3", "firm4", "firm5"];
market = [20, 25, 10, 15, 20];
plt.pie(market, labels=firms);
plt.axis('equal');
plt.legend(title='List of Firms');
plt.show();
```

```python
tips_df = sns.load_dataset('tips');
tips_df.head()
```

```python
tips_df["tips_percentage"] = tips_df["tip"] / tips_df["total_bill"];
tips_df.head()
```

```python
sns.histplot(tips_df["tips_percentage"], binwidth=0.05)
```

```python
sns.histplot(data=tips_df, x = "tips_percentage", binwidth=0.05, hue='time')
```

```python
sns.barplot(data=tips_df, x ='day', y='tip', estimator=np.sum)
```

```python
pivot = tips_df.pivot_table(
    index=['day'],
    columns=['size'],
    values=['tips_percentage'],
    aggfunc=np.average
)
sns.heatmap(pivot)
```
