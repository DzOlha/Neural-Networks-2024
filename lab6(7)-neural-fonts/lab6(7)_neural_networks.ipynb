## 1) Завантаження даних
Я обрала такі шрифти для побудови мережі: Arial, Book, Cambria, Italic, Lucida, Monotxt, Ocrb, Segoe, Times.

Вони містяться у архіві, доданому до завдання.

```python
from google.colab import files
import zipfile
import os
import pandas as pd
import glob
import shutil  # Import shutil for removing directories

# Step 1: Define the dataset directory and clear it if it exists
dataset_dir = '/content/dataset/'

# Clear the dataset directory
if os.path.exists(dataset_dir):
    shutil.rmtree(dataset_dir)  # Remove the entire directory and its contents

# Create the dataset directory again
os.makedirs(dataset_dir, exist_ok=True)
```

```python
# Step 2: Upload the zipped folder
uploaded = files.upload()
```

```python
# Step 3: Unzip the folder
for filename in uploaded.keys():
    if filename.endswith('.zip'):
        zip_ref = zipfile.ZipFile(filename, 'r')
        zip_ref.extractall(dataset_dir)  # Extract to /content/dataset folder
        zip_ref.close()

# Step 4: List all CSV files in the dataset directory and its subdirectories
csv_files = glob.glob(f'{dataset_dir}/**/*.csv', recursive=True)  # Search recursively
print(f"Found {len(csv_files)} CSV files.")

# Step 5: Merge all CSV files into a single DataFrame
dataframes = []
for csv_file in csv_files:
    print(f"Loading {csv_file}...")
    df = pd.read_csv(csv_file)  # Load each CSV file into a DataFrame
    dataframes.append(df)
```

```python
# Step 6: Check if dataframes list is not empty before concatenating
if dataframes:
    merged_df = pd.concat(dataframes, ignore_index=True)

    # Step 7: Save the merged DataFrame to a CSV file
    merged_df.to_csv(os.path.join(dataset_dir, 'merged_fonts.csv'), index=False)
    print("Merged dataset saved as 'merged_fonts.csv'")
else:
    print("No CSV files were found to merge.")
```

## 2) Попередня підготовка

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Step 1: Separate the target and features
X = merged_df.drop(columns=['font', 'fontVariant'])  # Drop both target and duplicate column
y = merged_df['font']  # Target column (font)

# Step 2: Define numerical attributes (all remaining features are numerical)
num_attribs = ['strength', 'italic', 'orientation', 'm_top', 'm_left', 'originalH', 'originalW'] + [f'r19c{i}' for i in range(10, 20)]

# Step 3: Create Pipelines for Numerical Features
# Numerical pipeline - imputing missing values and scaling
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),  # Fill missing values with median
    ('scaler', StandardScaler())  # Standardize features
])

# Step 4: Combine Pipelines using ColumnTransformer (since only numerical data, no categorical pipeline is needed)
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs)
])

# Step 5: Fit and Transform the Features
X_prepared = full_pipeline.fit_transform(X)

# Step 6: One-Hot Encode the Target Variable (y)
y_onehot = OneHotEncoder(sparse_output=False)  # Set sparse_output=False to return a dense array
y_encoded = y_onehot.fit_transform(y.values.reshape(-1, 1))  # Reshape y for encoding

# Step 7: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_prepared, y_encoded, test_size=0.2, random_state=42)
```

## 3) Побудова нейронної мережі

```python
!pip install scikeras

from scikeras.wrappers import KerasClassifier

import numpy as np
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


# Step 1: Define a function to create the model with variable layers and neurons
def create_model(num_hidden_layers=1, num_neurons=128, optimizer='adam', dropout_rate=0.3):
    model = Sequential()

    # Input layer
    model.add(Dense(num_neurons, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dropout(dropout_rate))  # Dropout to prevent overfitting

    # Adding hidden layers based on num_hidden_layers
    for _ in range(num_hidden_layers - 1):
        model.add(Dense(num_neurons, activation='relu'))
        model.add(Dropout(dropout_rate))  # Dropout for each hidden layer

    # Output layer
    model.add(Dense(y_train.shape[1], activation='softmax'))  # Adjust this for multi-class classification
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

## 4) Підбір гаперпараметрів

```python
# Step 2: Wrap the model using KerasClassifier
model = KerasClassifier(build_fn=create_model, verbose=0, dropout_rate=0.3, num_hidden_layers=1, num_neurons=128)
```

```python
# Step 3: Define the grid of hyperparameters to search
param_grid = {
    'num_hidden_layers': [2, 3],         # Experiment with 2 and 3 hidden layers
    'batch_size': [32, 64],              # Different batch sizes
    'epochs': [30, 40]                     # Number of epochs to train the model
}
```

```python
# Step 4: Setup GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=2)  # Use 2-fold cross-validation
```

```python
# Step 5: Fit GridSearchCV
grid_result = grid.fit(X_train, y_train)

# Step 6: Print the best parameters and score
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")

# Get the results for all combinations
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

for mean, std, param in zip(means, stds, params):
    print(f"{mean} (+/- {std}) with: {param}")
```

```python
# Step 7: Retrieve the best parameters from grid search
#best_params = grid_result.best_params_
best_params = {
    'num_hidden_layers': 3,
    'batch_size': 64,
    'epochs': 40
}
print(f"Best Parameters: {best_params}")

# Step 8: Create a new model using the best parameters
best_model = create_model(
    num_hidden_layers=best_params['num_hidden_layers']
)
```

```python
# Step 9: Train the model with the training data
best_model.fit(X_train, y_train, batch_size=best_params['batch_size'], epochs=best_params['epochs'], verbose=1)
```

```python
# Step 10: Evaluate the model on the test data
test_loss, test_accuracy = best_model.evaluate(X_test, y_test, verbose=1)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Step 11: Save the model
best_model.save('font_classification_model.keras')
```
